# Vectors

```python
# sum of two vectors
def addn(v,w): return [v[i]+w[i] for i in range(len(w))]
addn([0,1,2], [0,1,2])
```

```python
# product of alpha and a vector
def scalar_vector_mult(alpha, v): return [alpha*v[i] for i in range(len(v))]
scalar_vector_mult(2,[0,1])
```

```python
class Vec:
    def __init__(self, labels, function):
        self.D = labels    #set
        self.f = function  #represented as dict
        
    def setitem(self, d, val):
        f[d] = val
    
    def getitem(self, d): 
        return self.f[d] if d in self.f else 0.0
```

```python
v = Vec({'A','B','C'},{'A':1.0})
for d in v.D:
    if d in v.f:
        print(v.f[d])

print(v.getitem('A'))
print(v.getitem('B'))
```

```python
def no_mapping(v): return v.D-v.f.keys()
D0 = no_mapping(v)
print(D0)
```

```python
# Zero-vector
def zero_vec(D): return Vec(D,{})
v0 = zero_vec({0,1})
print(v0.D, v0.f)
```
